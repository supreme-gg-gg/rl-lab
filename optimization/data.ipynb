```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Investment Universum (Yahoo Finance symbols)
sym = [
    'AAPL',      # Apple Inc. Stock
    # 'AMZN',    # Amazon Inc. Stock (commented out for now)
    '^GSPC',     # S&P 500 Index
    'GC=F',      # Gold Futures
    'BTC-USD'    # Bitcoin in USD
]

# Download data using yfinance
data = yf.download(sym, start='2018-01-01', auto_adjust=True)

data = data['Close']

# Drop any missing values
data.dropna(inplace=True)

# Plot each asset's prices in individual subplots
fig, axes = plt.subplots(len(data.columns), 1, figsize=(10, 8), sharex=True)

for i, col in enumerate(data.columns):
    axes[i].plot(data.index, data[col])
    axes[i].set_title(col)
    axes[i].grid(True)

# Display the plots
plt.tight_layout()
plt.show()

# Plot normalized prices for comparison
plt.figure(figsize=(10, 6))
normalized_data = data / data.iloc[0]
plt.plot(normalized_data)
plt.title('Normalized Asset Prices')
plt.legend(normalized_data.columns)
plt.grid(True)
plt.show()
```

```python
data.to_csv('universe.csv')
```
